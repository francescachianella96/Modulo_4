```python
import numpy as np
```

1. Inizializzazione

```python
# 3 feature in ingresso e 5 neuroni nello strato nascosto
input_dim = 3
hidden_dim = 5

# iniziallizazione dei pesi molto piccoli e diversi (MAI A ZERO)
W = np.random.randn(hidden_dim, input_dim) * 0.01

# definizione di bias
b = np.zeros((hidden_dim, 1))

W, b
```

2. Esecuzione del calcolo

```python
# simulazione dato in ingresso
x = np.random.randn(input_dim, 1)

# calcolo del segnale
z = np.dot(W, x) + b

z
```

```python
# passaggio dei dati in batch
batch_size = 4
X_batch = np.random.randn(batch_size, input_dim)

# calcolo vettorializzato, attenzione a trasformare l'array per la moltiplicazione con la matrice

Z_batch = np.dot(W, X_batch.T) + b

Z_batch
```

```python
# generazione dei persi con xavier per il calcolo della varianza corretta in base al numero delle feature

def initialize_xavier(n_in, n_out):
# La formula di Xavier: varianza bilanciata tra input e output
# std = sqrt(2 / (n_in + n_out))\n",
    limit = np.sqrt(2.0 / (n_in + n_out))

# Generiamo pesi con questa specifica scala
    W_x = np.random.randn(n_out, n_in) * limit
    b_x = np.zeros((n_out, 1))

    return W_x, b_x

# Testiamo la nuova inizializzazione
W_new, b_new = initialize_xavier(input_dim, hidden_dim)
print(f"Nuova deviazione standard dei pesi: {np.std(W_new):.4f}")
```
